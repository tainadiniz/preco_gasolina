```python
import matplotlib.pyplot as plt
import seaborn as sns
import unidecode
#Leitura dos dados 
url = "https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/shpc/dsan/2024/precos-gasolina-etanol-08.csv"

data = pd.read_csv(url, delimiter=";").rename(columns=lambda col: unidecode.unidecode(col).strip().lower().replace(' ', '_'))
data = data.rename(columns={
    'regiao_-_sigla': 'regiao_sigla',
    'estado_-_sigla': 'estado_sigla'
})                                            
data.head()
```

```python
#Dados da Paraíba
data_pb = data[data['estado_sigla'] == 'PB']
data_pb['valor_de_venda'] = data_pb['valor_de_venda'].str.replace(',', '.').astype(float)
data_pb
```

```python
preco_medio_municipio = data_pb.groupby(['estado_sigla', 'municipio',"bairro", 'produto'])['valor_de_venda'].mean().reset_index()
preco_medio_municipio = preco_medio_municipio.sort_values(by='valor_de_venda', ascending=False)
preco_medio_municipio_joao_pessoa = preco_medio_municipio[preco_medio_municipio['municipio'] == 'JOAO PESSOA']
preco_medio_municipio_joao_pessoa
```

```python
# Filtrando por Gasolina
preco_medio_gasolina = preco_medio_municipio_joao_pessoa[preco_medio_municipio_joao_pessoa['produto'] == 'GASOLINA']

plt.figure(figsize=(10,6))  
plt.bar(preco_medio_gasolina['bairro'], preco_medio_gasolina['valor_de_venda'], color='green')

for i, valor in enumerate(preco_medio_gasolina['valor_de_venda']):
    plt.text(i, valor + 0.01, f'{valor:.2f}', ha='center')

plt.xticks(rotation=45, ha='right')
plt.title('Preços Médios de Gasolina por Bairro - João Pessoa', fontsize=14)
plt.xlabel('Bairro', fontsize=12)
plt.ylabel('Preço Médio (R$)', fontsize=12)

plt.tight_layout()
plt.show()
```
